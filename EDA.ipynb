```python
import pandas as pd
import numpy as np
p='https://raw.githubusercontent.com/nunnarilabs/ml/master/Automobile_Data/automobile_data.csv'
data=pd.read_csv(p)
data
```

```python
data.replace("?",np.nan,inplace=True)
data.head(5)
data
data.isnull().sum()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.drop_duplicates(inplace=True)
data
```

```python
avg_norm_loss=data["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized",avg_norm_loss)
```

```python
data["normalized-losses"].replace(np.nan,avg_norm_loss,inplace=True)
data["bore"].replace(np.nan,data["bore"].astype("float").mean(axis=0),inplace=True)
```

```python
data["stroke"].replace(np.nan,data["stroke"].astype("float").mean(axis=0),inplace=True)
data["num-of-doors"].replace(np.nan,'four',inplace=True)
data["horsepower"].replace(np.nan,data["horsepower"].astype("float").mean(axis=0),inplace=True)
data["peak-rpm"].replace(np.nan,data["peak-rpm"].astype("float").mean(axis=0),inplace=True)
data["price"].replace(np.nan,data["price"].astype("float").mean(axis=0),inplace=True)
data.isnull().sum()
```

```python
data["num-of-doors"].mode()
```

```python
data.dropna(subset=["price"],axis=0,inplace=True)
data.reset_index(drop=True,inplace=True)
data.shape
```

```python
data[["normalized-losses"]]=data[["normalized-losses"]].astype("int")
data[["price","peak-rpm","horsepower","bore","stroke"]]=data[["price","peak-rpm","horsepower","bore","stroke"]].astype("float")
data
```

to identify the outliers
Z-score is called a standard score.this value help to understand that how far is the data point from the mean.And after setting up a threshold value one can
utilize z-score vzlue to define the outliers

```python
from scipy import stats
hp_zscore=np.abs(stats.zscore(data['normalized-losses']))
hp_zscore
```

```python
threshold=hp_zscore.max()
index_val=np.where(hp_zscore>=threshold)
data.iloc[index_val]['normalized-losses']
```

```python
new_data=data.drop(index=data.iloc[index_val]['normalized-losses'].index)
new_data
```

```python
q1=data['length'].quantile(0.25)
q3=data['length'].quantile(0.75)
IQR=q3-q1
lower=q1-1.5*IQR
upper=q3+1.5*IQR
upper_array=np.where(data['length']>=upper)[0]
lower_array=np.where(data['length']<=lower)[0]
```
